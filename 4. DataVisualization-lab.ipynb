## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0, 0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
df = pd.read_sql_query("SELECT * FROM master", conn)
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Annual Compensation')
plt.xlabel('Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

plt.boxplot(df['ConvertedComp'].dropna())
plt.title('Box Plot of Compensation')
plt.ylabel('Compensation')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], color='skyblue', alpha=0.5)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

bubble_size = df['Age'] * 10

plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, color='skyblue', alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

query = "SELECT * FROM DatabaseDesireNextYear"
df = pd.read_sql_query(query, conn)

database_counts = df['DatabaseDesireNextYear'].value_counts().head(5)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(database_counts, labels=database_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df = pd.read_sql_query("SELECT * FROM master", conn)

df_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

grouped = df_filtered.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

grouped.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.legend(title='Hours', loc='upper left')
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df = pd.read_sql_query("SELECT * FROM master", conn)

df_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

grouped = df_filtered.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 6))
plt.plot(grouped.index, grouped.values, marker='o', color='skyblue', linestyle='-')
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()
```
