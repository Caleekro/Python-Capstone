## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here

plt.figure(figsize=(10, 6))
plt.boxplot(df['ConvertedComp'].dropna())
plt.title('Box Plot of Converted Compensation')
plt.ylabel('Converted Compensation')
plt.grid(True)
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
print("The Inter Quartile Range for the column ConvertedComp is:", 26868, "TO",  100000)
```

Find out the upper and lower bounds.


```python
# your code goes here

Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)

IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

print("Upper bound:", upper_bound)
print("Lower bound:", lower_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here

Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)

IQR = Q3 - Q1

lower_threshold = Q1 - 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR

outliers = df[(df['ConvertedComp'] < lower_threshold) | (df['ConvertedComp'] > upper_threshold)]

num_outliers = outliers.shape[0]

print("Number of outliers:", num_outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_filtered = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
df_filtered['ConvertedComp'].describe()
```
